```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('BONUS Track/alphabet_stock_data__1_.csv')
```

```python
df.head()
```

```python
df1 = df[(df['Date'] > '2020-04-01') & (df['Date'] <= '2020-10-30')]
x = df1["Date"]
y = df1["Adj Close"]
plt.figure(figsize=(10, 8))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.plot(x,y)
plt.title("Stock Prices Alphabet Inc. From 2020-04-01 To 2020-10-30")
plt.ylabel("Price")
plt.xlabel("Date")

```

```python

df2 = df[(df['Date'] > '2020-04-01') & (df['Date'] <= '2020-09-30')]
plt.figure(figsize=(10, 8))
lab = ['Open','High','Low','Close']
plt.hist([df2['Open'],df2['High'],df2['Low'],df2['Close']],bins=len(df2["Date"]), stacked=True, label=  lab)
plt.title("Opening/Closing/High/Low Stock Prices Alphabet Inc. From 2020-04-01 To 2020-09-30")
plt.ylabel("Frequency")
plt.xlabel("Date")
plt.legend()

```
